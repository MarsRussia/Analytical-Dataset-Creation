```python
import pandas as pd
import numpy as np
import datetime as dt
import os
```

```python
#choc_mapping = pd.read_excel("../Data/MappingFile/Choco_NielsenPOS_Hypercom_mapping.xlsx")
gums_mapping = pd.read_excel("../Data/MappingFile/GUMS Mapping Hypercom,nielsen,pos,POS_monthly color coding_ver0.7.xlsx")
hypercom_df = pd.read_excel("../Data/merged Hypercom 2017-2020.xlsx")

###SKUs needed to be extracted.
#gums_skus = pd.read_excel("../data/Chocos and Gums PPG for Data Extraction(Hypercom).xlsx",sheet_name = "GUMS PG Orbit OTC and Orbit XXl")
#chocos_skus = pd.read_excel("../data/Chocos and Gums PPG for Data Extraction(Hypercom).xlsx",sheet_name = "Chocos PG PPG_1,PPG_2,PPG_3")
gums_skus = pd.read_excel("../Data/Chocos and Gums PPG for Data Extraction(Hypercom).xlsx",sheet_name = "GUMS")
chocos_skus = pd.read_excel("../Data/Chocos and Gums PPG for Data Extraction(Hypercom).xlsx",sheet_name = "Choco")
```

```python
#GUMS
#new_df = pd.merge(gums_skus,gums_mapping, left_on="ID",right_on="ID",how='left')
#Chocos
#chocos_skus["ZREP CODE"] = chocos_skus["ZREP CODE"].astype(str) + '_0125' 
#new_df2 = pd.merge(chocos_skus,choc_mapping,left_on=['SKU', 'ZREP CODE'],right_on=["POS X5_SKU.Mars","POS X5_ZREP"])
choc_mapping = pd.read_excel("../Data/MappingFile/Choco_NielsenPOS_Hypercom_mapping.xlsx",skiprows=2)
choc_mapping = choc_mapping.drop("Remarks",axis=1)
choc_mapping
```

```python
#GUMS
new_df = pd.merge(gums_skus,gums_mapping, left_on="Nielsen_SKU",right_on="Nielsen_SKU",how='left')
#Chocos
new_df2 = pd.merge(chocos_skus,choc_mapping,left_on=["Nielsen_SKU"],right_on=["Nielsen_SKU"])
```

```python
new = new_df.filter(regex='Hypercom')
gum_hypercom = new_df.filter(regex='Russian')
```

```python
new=new_df2.filter(regex="Hypercom")
new = new[new.columns.drop(list(new.filter(regex='_ID')))]
choco_hypercom = new[new.columns.drop(list(new.filter(regex='_English')))]
```

```python
gum_hypercom.columns = choco_hypercom.columns
ref_hypercom = pd.concat([choco_hypercom,gum_hypercom])
ref = ref_hypercom.drop_duplicates()
```

```python
ref = ref.dropna(axis=0,how='all')
```

```python
res_df = pd.merge(ref,hypercom_df,left_on=['Hypercom_Brand_New', 'Hypercom_Position_New', 'Hypercom_Packaging','Hypercom_manufacturer', 'Hypercom_SEGMENT_OLD'],right_on= ['Brand_New', 'Position_New', 'Packaging','manufacturer', 'SEGMENT_OLD'], how='left')
```

```python
res_df = res_df[res_df.columns.drop(list(res_df.filter(regex='Hypercom_')))]
```

```python
rdf_gums = pd.merge(new_df,res_df,left_on=['Hypercom_Brand_New_Russian','Hypercom_Position_New_Russian','Hypercom_Packaging_Russian','Hypercom_manufacturer_Russian','Hypercom_SEGMENT_OLD_Russian'],right_on=['Brand_New','Position_New', 'Packaging', 'manufacturer', 'SEGMENT_OLD'],how='left')
rdf_choco = pd.merge(new_df2,res_df,left_on=['Hypercom_Brand_New','Hypercom_Position_New','Hypercom_Packaging','Hypercom_manufacturer','Hypercom_SEGMENT_OLD'],right_on=['Brand_New','Position_New', 'Packaging', 'manufacturer', 'SEGMENT_OLD'],how='left')
```

```python
#rdf_gums.columns
```

```python
Selecting_Columns = ["ID","Nielsen_Manufacturer","Nielsen_Brand","Nielsen_SKU","Seller","Format",
                      "Hypercom_Brand_New_English","Hypercom_Position_New_English","Hypercom_Packaging_English",
                      "Hypercom_manufacturer_English", "Hypercom_SEGMENT_OLD_English",
                      "start date","expiration date", "Headline","Catalog distribution",
                      "% Directory distribution","Distribution Promo","Distribution Promo%","Trade","Is free",
                      "Price without stock for goods","Promo Price","Price per unit","Virtual price","Discount depth","number of pages","number mage"]
rdf_gum = rdf_gums[Selecting_Columns]

Selecting_Columns = ["Nielsen_POS_ID","Nielsen_Manufacturer","Nielsen_Brand","Nielsen_SKU",
                      "POS X5_Segment","POS X5_Category","POS X5_ZREP","POS X5_SKU.Mars","POS X5_SKU","POS Tander_GRD.code...unique",
                      "POS Tander_Brand","POS Tander_Corporate.Segment",
                      "POS Tander_Product.Name","Seller","Format",
                      "Hypercom_Brand_New_English","Hypercom_Position_New_English","Hypercom_Packaging_English",
                      "Hypercom_manufacturer_English", "Hypercom_SEGMENT_OLD_English",
                      "start date","expiration date", "Headline","Catalog distribution",
                      "% Directory distribution","Distribution Promo","Distribution Promo%","Trade","Is free",
                      "Price without stock for goods","Promo Price","Price per unit","Virtual price","Discount depth","number of pages","number mage"]

rdf_choc = rdf_choco[Selecting_Columns]
```

```python
ppg_rdf_choc = pd.merge(rdf_choc,chocos_skus,right_on="Nielsen_SKU",left_on="Nielsen_SKU", how='left')
ppg_rdf_gum = pd.merge(rdf_gum,gums_skus,right_on="Nielsen_SKU",left_on="Nielsen_SKU",how='left')
```
