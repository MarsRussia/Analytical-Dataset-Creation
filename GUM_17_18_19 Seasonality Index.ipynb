```python
import pandas as  pd
import numpy as np
import datetime as dt
import os
os.getcwd()
```

```python
si_gum = pd.read_csv("../Data/Raw Data/(GUMS )Neilsen_Complete_data_long_format.csv")
```

```python
si_gum = si_gum.rename(columns={"1:Value (in 1000 RUR)": "value_sales"})
```

```python
si = pd.DataFrame()
si = si_gum[["Week","value_sales"]]
```

```python
si = si.groupby(by = "Week", as_index=False).sum()
```

```python
#for 2017 and 2018 and 2019
```

```python
si_17_18_19 = si
si_17_18_19.loc[155]
```

```python
si_17_18_19['Week'].loc[148] = 'W 2017 01'
si_17_18_19['Week'].loc[149] = 'W 2017 02'
si_17_18_19['Week'].loc[150] = 'W 2017 03'
si_17_18_19['Week'].loc[151] = 'W 2017 04'
si_17_18_19['Week'].loc[152] = 'W 2017 05'
si_17_18_19['Week'].loc[153] = 'W 2017 06'
si_17_18_19['Week'].loc[154] = 'W 2017 07'
si_17_18_19['Week'].loc[155] = 'W 2017 08'
```

```python
si_17_18_19['Year'] = si_17_18_19['Week'].str.slice(2, 6)
```

```python
si_17_18_19['week_number'] = si_17_18_19['Week'].str.slice(7, 9)
si_17_18_19
```

```python
si_17_18_19.Year = pd.to_numeric(si_17_18_19.Year)
si_17_18_19.week_number = pd.to_numeric(si_17_18_19.week_number)
```

```python
si_17_18_19 = si_17_18_19.loc[(si_17_18_19["Year"]==2017) | (si_17_18_19["Year"]==2018)  | (si_17_18_19["Year"]==2019)]
```

```python
si_17_18_19
```

```python
mysi = pd.DataFrame()
```

```python
mysi['week_number'] = np.arange(1,53,1)
```

```python
value_sales_2017=[]
value_sales_2018=[]
value_sales_2019=[]
for i in mysi.week_number:
    value_sales_2017.append(float(si_17_18_19.value_sales[(si_17_18_19.Year==2017) & (si_17_18_19.week_number==i)]))
    value_sales_2018.append(float(si_17_18_19.value_sales[(si_17_18_19.Year==2018) & (si_17_18_19.week_number==i)]))
    value_sales_2019.append(float(si_17_18_19.value_sales[(si_17_18_19.Year==2019) & (si_17_18_19.week_number==i)]))
```

```python
mysi['value_sales_2017'] = value_sales_2017
mysi['value_sales_2018'] = value_sales_2018
mysi['value_sales_2019'] = value_sales_2019
mysi
```

```python
mysi.value_sales_2017 = mysi.value_sales_2017/np.mean(mysi.value_sales_2017)
mysi.value_sales_2018 = mysi.value_sales_2018/np.mean(mysi.value_sales_2018)
mysi.value_sales_2019 = mysi.value_sales_2019/np.mean(mysi.value_sales_2019)
mysi
```

```python
mysi['SI'] = (mysi.value_sales_2017 + mysi.value_sales_2018 + mysi.value_sales_2019)/3
```

```python
del mysi["value_sales_2017"]
del mysi["value_sales_2018"]
del mysi["value_sales_2019"]
mysi
```
