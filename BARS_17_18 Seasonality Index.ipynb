```python
import pandas as  pd
import numpy as np
import datetime as dt
import os
os.getcwd()
```

```python
si_choco = pd.read_csv(r"../Data/Raw Data/(BARS)MERGED 2017-2020 NIELSEN BARS DATA LONG FORMAT.csv")
si_choco.shape
```

```python
si_choco = si_choco[~si_choco.SKU.str.contains("TOTAL IMPULSE CHOCOLATE")]
si_choco.shape
```

```python
si_choco.columns
```

```python
si_choco = si_choco.rename(columns={"1:Value (in 1000 RUR)": "value_sales"})
```

```python
si = pd.DataFrame()
si = si_choco[["Week","value_sales"]]
```

```python
si = si.groupby(by = "Week", as_index=False).sum()
```

```python
#for 2017 and 2018
```

```python
si_17_18 = si
si_17_18.loc[96]
```

```python
si_17_18['Week'].loc[96] = 'W 2017 01'
si_17_18['Week'].loc[97] = 'W 2017 02'
si_17_18['Week'].loc[98] = 'W 2017 03'
si_17_18['Week'].loc[99] = 'W 2017 04'
si_17_18['Week'].loc[100] = 'W 2017 05'
si_17_18['Week'].loc[101] = 'W 2017 06'
si_17_18['Week'].loc[102] = 'W 2017 07'
si_17_18['Week'].loc[103] = 'W 2017 08'
```

```python
si_17_18['Year'] = si_17_18['Week'].str.slice(2, 6)
```

```python
si_17_18['week_number'] = si_17_18['Week'].str.slice(7, 9)
si_17_18
```

```python
si_17_18.Year = pd.to_numeric(si_17_18.Year)
si_17_18.week_number = pd.to_numeric(si_17_18.week_number)
```

```python
si_17_18 = si_17_18.loc[(si_17_18["Year"]==2017) | (si_17_18["Year"]==2018)]
```

```python
si_17_18
```

```python
mysi = pd.DataFrame()
```

```python
mysi['week_number'] = np.arange(1,53,1)
```

```python
value_sales_2017=[]
value_sales_2018=[]
for i in mysi.week_number:
    value_sales_2017.append(float(si_17_18.value_sales[(si_17_18.Year==2017) & (si_17_18.week_number==i)]))
    value_sales_2018.append(float(si_17_18.value_sales[(si_17_18.Year==2018) & (si_17_18.week_number==i)]))
```

```python
mysi['value_sales_2017'] = value_sales_2017
mysi['value_sales_2018'] = value_sales_2018
```

```python
mysi.value_sales_2017 = mysi.value_sales_2017/np.mean(mysi.value_sales_2017)
mysi.value_sales_2018 = mysi.value_sales_2018/np.mean(mysi.value_sales_2018)
```

```python
mysi['SI'] = (mysi.value_sales_2017 + mysi.value_sales_2018)/2
```

```python
del mysi["value_sales_2017"]
del mysi["value_sales_2018"]
mysi
```

```python
mysi.to_excel(r"../Process/SI_Index/Bars_17_18 Seasonality Index.xlsx", index=False)
```
